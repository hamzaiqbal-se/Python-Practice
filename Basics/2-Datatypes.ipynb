Basic Data Types

```python
# integer
print(5)
print(1e308)  # maximu range of integer is 1*10 to the power 308 
print(1e309)
```

```python
# float
print(5.1)
print(1.7e308)  # maximu range of float is 1.7*10 to the power 308 
```

```python
# boolean
print(True)
print(False)
```

```python
# String
print('hamza')
print("hamza")
print('''hamza''')
print('5')
```

```python
# complex
print(4+5j)
```

Container Types

```python
# List
print([1,2,3,4,5])
```

```python
# Tuple
print((1,2,3,4,5))
```

```python
# Set
print({1,2,3,4,5})
```

```python
# Dict
print({"Name":"Hamza","Age":22,"Gender":"Male"})
```
