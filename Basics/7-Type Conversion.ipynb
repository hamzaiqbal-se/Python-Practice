Type conversion is not a permanent operation.
It creates new integer, that means there is no change in original value

```python
a = 4.5
int(a)  # type conversion
```

```python
print(a)   # value is still 4.5
```

```python
num1= input("Enter 1st number: ")
num2= input("Enter 2nd number: ")
add = int(num1) + int(num2)         # type conversion
print('Addition of both numbers is: ', add)
```

```python
# Better Approach
num1= int(input("Enter 1st number: "))
num2= int(input("Enter 2nd number: "))
add = num1 + num2
print('Addition of both numbers is: ', add)
```
